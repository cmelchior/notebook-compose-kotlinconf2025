```kotlin
import java.io.File

val json = File("/Users/christian.melchior/colors.json").readText()
json
```

```kotlin
%use serialization
```

```kotlin
val jsonColors = json.deserializeJson()
jsonColors
```

```kotlin
data class ColorData(val title: String, val color: Color)

val composeColors = jsonColors.cssColors.map {
    ColorData(it.name, Color(it.long))
}
composeColors
```

```kotlin
@Composable
fun SimpleBox() {
    val (title, color) = composeColors.first()
    Box(
        modifier = Modifier.background(color = color).size(100.dp),
        contentAlignment = Alignment.Center
    ) {
        Text(text = title)
    }
}

COMPOSE(backgroundColor = Color.White) {
    SimpleBox()
}
```

```kotlin
@file:OptIn(ExperimentalLayoutApi::class)

import com.example.toHsv

val sortedColors = composeColors.sortedBy { it.color.toHsv() }

@Composable
fun ColorBox(
    data: ColorData,
    onSelected: () -> Unit = {}
) {
    Box(
        modifier = Modifier
            .clip(RoundedCornerShape(8.dp))
            .clickable { onSelected() }
            .background(color = data.color)
            .size(60.dp)
            .padding(4.dp),
        contentAlignment = Alignment.Center
    ) {
        Text(text = data.title, fontSize = 10.sp, maxLines = 1, overflow = TextOverflow.Ellipsis)
    }
}


@Composable
fun AllColors(startData: ColorData = ColorData("No Color", Color.Transparent)) {
    var selectedColor by remember { mutableStateOf<ColorData>(startData) }
    Column(modifier = Modifier.padding(24.dp)) {
        Box(modifier = Modifier
            .fillMaxWidth()
            .padding(bottom = 24.dp),
            contentAlignment = Alignment.Center
        ) {
            ProjectDialog(selectedColor.title, selectedColor.color)
        }
        FlowRow(
            horizontalArrangement = Arrangement.spacedBy(8.dp),
            verticalArrangement = Arrangement.spacedBy(8.dp),
        ) {
            sortedColors.forEach { colorData  ->
                ColorBox(data = colorData) { selectedColor = colorData }
            }
        }
    }
}

COMPOSE(height = 1000) {
    AllColors()
}
```

```kotlin
// Launch applications directly from the Notebook
// But set `exitProcessOnExit = false` to avoid IDEA closing as well
application(exitProcessOnExit = false) {
    Window(onCloseRequest = ::exitApplication, title = "Compose Color Selector") {
        AllColors()
    }
}
```
